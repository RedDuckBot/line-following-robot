```python
import cv2, os 
import numpy as np
import matplotlib.pyplot as plt
```

This notebook covers how the yellow pedals of a sunflower can be tracked or masked out from an image.

```python
root_dir = os.getcwd()
img_path = os.path.join(root_dir,"sunFlower.jpg")
```

```python
original_img = cv2.imread(img_path)
imgRGB = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
hsv_img = cv2.cvtColor(original_img,cv2.COLOR_BGR2HSV)

lowerBound = np.array([19,215,215])
upperBound = np.array([27,255,255])

mask = cv2.inRange(hsv_img,lowerBound,upperBound)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 5))

ax[0][0].imshow(original_img)
ax[0][0].axis('off')
ax[0][0].set_title("BGR flower")

ax[0][1].imshow(imgRGB)
ax[0][1].axis('off')
ax[0][1].set_title("RGB Flower")
ax[0][0].imshow(original_img)

ax[1][0].imshow(hsv_img)
ax[1][0].axis('off')
ax[1][0].set_title("HSV flower")

ax[1][1].imshow(mask)
ax[1][1].axis('off')
ax[1][1].set_title("mask")

```
